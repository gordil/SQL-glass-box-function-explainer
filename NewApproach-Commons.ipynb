```
# Function to validate order_id
def validate_order_id(input_vw):
    output_vw = "vw_validate_order_id"
    sql(f"""
        CREATE OR REPLACE TEMP VIEW {output_vw} AS
        SELECT  * EXCEPT (validity),
        CASE
            WHEN order_id <= 0 THEN 'Invalid Order ID'
            ELSE validity 
        END AS validity
        FROM {input_vw};
    """)
    return output_vw


# Function to validate quantity
def validate_quantity(input_vw):
    output_vw = "vw_validate_quantity"
    sql(f"""
        CREATE OR REPLACE TEMP VIEW {output_vw} AS
        SELECT  * EXCEPT (validity),
        CASE
            WHEN quantity <= 0 THEN 'Invalid Quantity'
            ELSE validity 
        END AS validity
        FROM {input_vw};
    """)
    return output_vw


# Function to validate phone_number
def validate_phone_number(input_vw):
    output_vw = "vw_validate_phone_number"
    sql(f"""
        CREATE OR REPLACE TEMP VIEW {output_vw} AS
        SELECT  * EXCEPT (validity),
        CASE
            WHEN LENGTH(phone_number) != 10 THEN 'Invalid Phone Number'
            ELSE validity 
        END AS validity
        FROM {input_vw};
    """)
    return output_vw


# Function to validate email
def validate_email(input_vw):
    output_vw = "vw_validate_email"
    sql(f"""
        CREATE OR REPLACE TEMP VIEW {output_vw} AS
        SELECT  * EXCEPT (validity),
        CASE
            WHEN customer_email IS NULL OR customer_email = '' THEN 'Missing Email'
            ELSE validity 
        END AS validity
        FROM {input_vw};
    """)
    return output_vw


# Function to validate order_date
def validate_order_date(input_vw):
    output_vw = "vw_validate_order_date"
    sql(f"""
        CREATE OR REPLACE TEMP VIEW {output_vw} AS
        SELECT  * EXCEPT (validity),
        CASE
            WHEN order_date > CURRENT_DATE THEN 'Order Date is in the Future'
            ELSE validity 
        END AS validity
        FROM {input_vw};
    """)
    return output_vw


# Function to validate discount_percentage
def validate_discount_percentage(input_vw):
    output_vw = "vw_validate_discount_percentage"
    sql(f"""
        CREATE OR REPLACE TEMP VIEW {output_vw} AS
        SELECT  * EXCEPT (validity),
        CASE
            WHEN discount_percentage < 0 OR discount_percentage > 50 THEN 'Invalid Discount Percentage'
            ELSE validity 
        END AS validity
        FROM {input_vw};
    """)
    return output_vw
```
